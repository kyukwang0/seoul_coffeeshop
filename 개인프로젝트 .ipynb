# 스타벅스를 기준으로 커피숍 브랜드별 현황

```python
coffe_shop= data[(data)["uptaenm"]=="커피숍"]                                   # 업태가 커피숍 인걸 모았음
```

커피숍 브랜드 데이터프레임

```python
star_df = pd.DataFrame()                                                        # 스타벅스 + 데이터프레임
ediya_df = pd.DataFrame()

# 각 커피숍에 넣을 데이터 프레임을 temp_df라는 이름으로 만들기
temp_df = pd.DataFrame(
    data={
        "dtlstatenm" : "",                                                      # 영업상태명
        "apvpermymd" : "",                                                      # 인허가일자
        "dcbymd" : "",                                                          # 페업일자
        "mgtno" : "",                                                           # 관리번호
        "x" : "",                                                               # x축
        "y" : "",                                                               # y축
        "rdnwhladdr" : ""                                                       # 도로명주소
    }, index = [0]
)
```

영업중인 스타벅스 매장 중 컬럼값으로 정리한 것

```python

for i in coffe_shop.index:
    temp = coffe_shop.at[i, "bplcnm"]                                           #커피숍데이터내에 이름을 temp로 저장, 
    temp1 = coffe_shop.at[i, "dtlstatenm"]                                      #커피숍데이터내에 영업여부를 temp로 저장, 
    if len(re.findall(r"스타벅스",temp))>=1 and len(re.findall(r'영업',temp1))>=1:# 스타벅스 이름이 들어간것 중에서
                                                                                 # 해당한는 컬럼값이 데이터를 찾는다
        temp_df["dtlstatenm"] = coffe_shop.at[i, "dtlstatenm"],  
        temp_df["apvpermymd"] = coffe_shop.at[i, "apvpermymd"],
        temp_df["dcbymd"] = coffe_shop.at[i, "dcbymd"],
        temp_df["mgtno"] = coffe_shop.at[i, "mgtno"],
        temp_df["x"] = coffe_shop.at[i, "x"],
        temp_df["y"] = coffe_shop.at[i, "y"],
        temp_df["rdnwhladdr"] = coffe_shop.at[i, "rdnwhladdr"]
        
        count = 0                                                               #중복된 숫자 카운트를 위해 count를 0으로
        if i >= 1:                                                              
            for j in star_df.index :                                            # 스타벅스 인덱스 값을 1개 쌓아주기
                if temp_df.at[0, "mgtno"] == star_df.at[j, "mgtno"]:            #기존 데이터와 스타벅스df내에 mgtno가 중복된다면
                    count += 1                                                  #카운트 에 1 더해주기
        if count == 0 :                                                         #카운트가 0일경우
             star_df = pd.concat([star_df, temp_df], ignore_index = True)       # 위에 찾은 컬럼내에 값을 concat으로 집어 넣어줌

    for j in temp_df.columns:                                                   #초기화코드
        temp_df.at[0, j] = ""                                                   #초기화코드


    
    if len(re.findall(r"이디야",temp))>=1 and len(re.findall(r'영업',temp1))>=1:# 영업중인 이디야 찾기
                                                                                 
        temp_df["dtlstatenm"] = coffe_shop.at[i, "dtlstatenm"],  
        temp_df["apvpermymd"] = coffe_shop.at[i, "apvpermymd"],
        temp_df["dcbymd"] = coffe_shop.at[i, "dcbymd"],
        temp_df["mgtno"] = coffe_shop.at[i, "mgtno"],
        temp_df["x"] = coffe_shop.at[i, "x"],
        temp_df["y"] = coffe_shop.at[i, "y"],
        temp_df["rdnwhladdr"] = coffe_shop.at[i, "rdnwhladdr"]

        count = 0
        if i >= 1:                                                              
            for j in ediya_df.index :  
                count = 0                                                      
                if temp_df.at[0, "mgtno"] == ediya_df.at[j, "mgtno"]:            
                    count += 1                                                  
        if count == 0 :                                                         
             ediya_df = pd.concat([ediya_df, temp_df], ignore_index = True)     


    for j in temp_df.columns:                                                   
        temp_df.at[0, j] = ""                                                   







```
